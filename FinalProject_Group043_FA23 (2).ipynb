# Abstract

Our group investigated the effect of green spaces on mental health because of the interest on mental health. Several studies have examined the effects of accessible green spaces and how it can play an effect on mental health. We took this conclusion one step forward to examine these effects on college campuses: how accessibility to green spaces and using green spaces play an effect on the mental health of college students. Our hypothesis states that green spaces do play an effect on mental health and the effect is positive: improving the mental health of students. To examine this effect, we used a dataset from an app called StudentLife that tracked several data such as physical activity and mental health questions. # Research Question

Is there a relationship between college student's amount of physical activity in terms of average time spent exercising and walking daily and their mental health measured through self-reported positive affect and stress levels to indirectly investigate the effect of access to urban green spaces on health?

```python
# Created a dictionary to match each response to a numerical value
answer_dict = {'Never': 0, 'Almost never': 1, 'Sometime': 2, 'Fairly often': 3, 'Very often': 4}
def stress_number(response):
    return answer_dict[response]
```

```python
df_stress_copied = df_stress.copy()
# df_stress_copied[df_stress_copied.columns[2:]] = df_stress_copied[
#     df_stress_copied.columns[2:]].apply(lambda x: stress_number(x), axis=0)
df_stress_copied = df_stress_copied.drop(columns = ['uid', 'type'])

for col in range(len(df_stress_copied.columns)):
    df_stress_copied[df_stress_copied.columns[col]] = df_stress_copied[df_stress_copied.columns[col]].apply(stress_number)

df_stress_copied
```

```python
df_stress_copied.insert(loc = 0, column = 'uid', value = df_stress_leftout['uid'])
df_stress_copied.insert(loc = 1, column = 'type', value = df_stress_leftout['type'])
```

### Spearman Correlation Test

In analyzing our data through testing, we first chose to do the Spearman Correlation Test for each relationship of interest with our variables. We chose this test as it's non-parametric, as some of our exercise data was not normally distirbuted as seen above. We also chose it as it enabled us to test the monotonic relationships between our ordinal and continuous variables. In other words, we could see the direction and strength of relationships between our variables of interest. 

```python
import scipy as scipy
```

```python
# correlation between down feelings and daily minutes walking
scipy.stats.spearmanr(a = ph_exercise['feeling hopeless and depressed'], b = ph_exercise['walking'])
```

- In comparing feelings of feelings of hopelessness and depression and minutes spent walking daily on average, we can see that the relationship is not strong or significant, as the statistic is 0.03 while our p-value, 0.84, is much larger than 0.05. As a result, we cannot reject the null therefore our data does not support a monotopic relationship between these two variables. 

```python
# correlation between down feelings and daily minutes exercising
scipy.stats.spearmanr(a = ph_exercise['feeling hopeless and depressed'], b = ph_exercise['exercise'])
```

- In comparing feelings of feelings of hopelessness and depression and minutes spent exercising daily on average, the resulting values show that the relationship is not strong or significant, as the test statistic is -0.017 while our p-value, 0.92, is significantly larger than 0.05. As a result, we cannot reject the null therefore our data does not support a monotopic relationship between these two variables as well. 

```python
# correlation between stress and daily minutes walking
scipy.stats.spearmanr(a = stress_exercise['feeling nervous and stressed'], b = stress_exercise['walking'])
```

- Next, we are looking at the relationship between percieved feelings of stress and minutes spent walking daily on average. Despite being better than the last couple tests, the resulting values once again show that the relationship is not strong or significant, with a test statistic of -0.07 while our p-value, 0.66, is greater than 0.05. As a result, we cannot reject the null therefore our data does not support a monotopic relationship between these two variables either. 

```python
# correlation between stress and daily minutes exercising
scipy.stats.spearmanr(a = stress_exercise['feeling nervous and stressed'], b = stress_exercise['exercise'])
```

- Finally, we are looking at the relationship between percieved feelings of stress and minutes spent exercising daily on average. This test showed the most potential, yet the resulting values once again show that the relationship is not strong or significant. The produced test statistic was 0.19 and our resulting p-value of 0.25 is still greater than 0.05. As a result, we cannot reject the null hypothesis. We can conclude that our data does not support a monotopic relationship between these two variables either, though it is intersting to note that the potential positive relationship seen through the statistic 0.19 does correspond with our original findings of stress and exercise levels increasing together in EDA. 

### Mann - Whitney U Test

As well, we chose to perform the Mann - Whitney U test with our data since this analysis works well with ordinal data. We chose this because it can work well with ordinal data and does not necessarily require the data to be normally distributed. Using this test, we were able to see the significance of our findings through the p-value produced and the statistic itself. This test not only works with ordinal data but can work with continuous data too.

```python
# comparing exercise and down feelings
import scipy.stats as stats
exercise = ph_exercise['exercise']
hopeless_dep = ph_exercise['feeling hopeless and depressed']
group1 = [exercise[i] for i in range(len(hopeless_dep)) if hopeless_dep[i] == 0]
group2 = [exercise[i] for i in range(len(hopeless_dep)) if hopeless_dep[i] == 3]
stats.mannwhitneyu(group1, group2)
```

- The test resulted in a statistic of 110.0 and a p-value of 0.15. Since the p value is above 0.05, we cannot reject the null hypothesis. This data shows the comparison between the distribution of the exercise data as they pertain to their corresponding response to "feeling hopeless and depressed". We examined the ordinal data of 0 and 1 as their responses, which correspond to not feeling hopeless and depressed with feeling very hopeless and depressed. This shows that there is little significant difference between the two groups, meaning that exercise did not impact the responses of these two groups

```python
walking = ph_exercise['walking']
hopeless_dep = ph_exercise['feeling hopeless and depressed']
group1 = [walking[i] for i in range(len(hopeless_dep)) if hopeless_dep[i] == 0]
group2 = [walking[i] for i in range(len(hopeless_dep)) if hopeless_dep[i] == 3]
stats.mannwhitneyu(group1, group2)
```

- This test resulted in a test statistic of 33.0 and pvalue of 0.34. Since the p value is above 0.05, we can't reject the null hypothesis and there is little significant difference between the two groups for their walking data based on their "feeling hopeless and depressed" data. We examined the lowest ordinal data value and the highest ordinal data value. Walking did not seem to impact the outcome of the feeling hopeless and depressed.
